```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#loading data from all 5 years
year2015 = pd.read_csv('../data/raw/2015.csv')
year2016 = pd.read_csv('../data/raw/2016.csv')
year2017 = pd.read_csv('../data/raw/2017.csv')
year2018 = pd.read_csv('../data/raw/2018.csv')
year2019 = pd.read_csv('../data/raw/2019.csv')
#columns of all 5 years data 
print('year2015' , '\n',year2015.columns , '\n')
print('year2016' , '\n',year2016.columns , '\n')
print('year2017' , '\n',year2017.columns , '\n')
print('year2018' , '\n',year2018.columns , '\n')
print('year2019' , '\n',year2019.columns , '\n')
```

```python
#selecting only the desired columns in ordered fashion
year2015 = year2015[[
                     'Country', 
                     'Happiness Rank', 
                     'Happiness Score', 
                     'Economy (GDP per Capita)', 
                     'Family', 
                     'Health (Life Expectancy)', 
                     'Freedom', 
                     'Trust (Government Corruption)', 
                     'Generosity'
                    ]]
year2016 = year2016[[
                    'Country',
                    'Happiness Rank',
                    'Happiness Score',
                    'Economy (GDP per Capita)',
                    'Family',
                    'Health (Life Expectancy)',
                    'Freedom',
                    'Trust (Government Corruption)',
                    'Generosity'
                    ]]
year2017 = year2017[[
                    'Country',
                    'Happiness.Rank',
                    'Happiness.Score',
                    'Economy..GDP.per.Capita.',
                    'Family',
                    'Health..Life.Expectancy.',
                    'Freedom',
                    'Trust..Government.Corruption.',
                    'Generosity'
                   ]]
year2018 = year2018[[
                    'Country or region',
                    'Overall rank',
                    'Score',
                    'GDP per capita',
                    'Social support',
                    'Healthy life expectancy',
                    'Freedom to make life choices',
                    'Perceptions of corruption',
                    'Generosity'
                   ]]
year2019 = year2019[[
                    'Country or region',
                    'Overall rank',
                    'Score',
                    'GDP per capita',
                    'Social support',
                    'Healthy life expectancy',
                    'Freedom to make life choices',
                    'Perceptions of corruption',
                    'Generosity'
                   ]]
#renaming columns to make every df uniform
rename={
       'Country':'Country',
       'Happiness Rank':'Happiness Rank',
       'Happiness score':'Happiness Score',
       'Economy (GDP per Capita)':'GDP/C',
       'Family':'Family',
       'Health (Life Expectancy)':'Health',
       'Freedom':'Freedom',
       'Trust (Government Corruption)':'Government Corruption',
       'Generosity':'Generosity'
}
rename2={
        'Country':'Country',
        'Happiness.Rank':'Happiness Rank',
        'Happiness.Score':'Happiness Score',
        'Economy..GDP.per.Capita.':'GDP/C',
        'Family':'Family',
        'Health..Life.Expectancy.':'Health',
        'Freedom':'Freedom',
        'Trust..Government.Corruption.':'Government Corruption',
        'Generosity':'Generosity'
}
rename3={
        'Country or region':'Country',
        'Overall rank':'Happiness Rank',
        'Score':'Happiness Score',
        'GDP per capita':'GDP/C',
        'Social support':'Family',
        'Healthy life expectancy':'Health',
        'Freedom to make life choices':'Freedom',
        'Perceptions of corruption':'Government Corruption',
        'Generosity':'Generosity'
}
year2015.rename(columns=rename, inplace=True)
year2016.rename(columns=rename, inplace=True)
year2017.rename(columns=rename2, inplace=True)
year2018.rename(columns=rename3, inplace=True)
year2019.rename(columns=rename3, inplace=True)

# adding year column
year2015['Year']=2015
year2016['Year']=2016
year2017['Year']=2017
year2018['Year']=2018
year2019['Year']=2019

#preview of the modified years data
print('year2015' , '\n' , year2015.head() , '\n')
print('year2016' , '\n' , year2016.head() , '\n')
print('year2017' , '\n' , year2017.head() , '\n')
print('year2018' , '\n' , year2018.head() , '\n')
print('year2019' , '\n' , year2019.head() , '\n')
```

```python
# combining all years into a singular dataframe 
data = pd.concat([year2019, year2018, year2017, year2016, year2015])
```

```python
#data cleaning

#checking for missing values
missing = data.isnull().sum()
print('number of missing values','\n' , missing)
print('\n')
#checking which row has missing value
missing_row = data[data['Government Corruption'].isnull()]
print('row containing missing value ','\n' , missing_row)

#filling missing value with average
data['Government Corruption'] = data['Government Corruption'].fillna(data['Government Corruption'].mean())
print('number of missing values after filling','\n',data['Government Corruption'].isnull().sum() , '\n')

#checking for duplicate values

#exact duplicates
print('number of duplicate rows ',data.duplicated().sum())
print('\n')
#duplicate contry in each ear data
print('number of duplicate country per year' , data.duplicated(subset=['Country' , 'Year']).sum())
print('\n')
#checking datatypes
print('data type of columns','\n',data.dtypes)

#standardizing names
data['Country'] = data['Country'].str.strip()
data['Country'] = data['Country'].str.lower()


#for renaming to reduce errors 
mapping = {
    "Taiwan Province of China": "Taiwan",
    "Taiwan": "Taiwan",
    "Hong Kong S.A.R., China": "Hong Kong",
    "Hong Kong": "Hong Kong",
    "Trinidad & Tobago": "Trinidad and Tobago",
    "Trinidad and Tobago": "Trinidad and Tobago",
    "Northern Cyprus": "Cyprus",
    "North Cyprus": "Cyprus",
    "Macedonia": "Macedonia",
    "North Macedonia": "Macedonia",
    "Somaliland region": "Somaliland Region",
    "Somaliland Region": "Somaliland Region",
    "Somalia": "Somalia",
    "Namibia": "Namibia",
    "Puerto Rico": "Puerto Rico",
    "Gambia": "Gambia",
    "Belize": "Belize",
    "South Sudan": "Sudan",
    "Mozambique": "Mozambique",
    "Lesotho": "Lesotho",
    "Oman": "Oman",
    "Swaziland": "Eswatini",
    "Djibouti": "Djibouti",
    "Central African Republic": "Central African Republic",
    "Laos": "Laos",
    "Suriname": "Suriname",
    "Comoros": "Comoros",
    "Angola": "Angola",
    "Sudan": "Sudan",
}
#renaming
data['Country'] = data['Country'].replace(mapping)

#calculating missing in each year
countries_2015 = data.loc[data['Year']==2015 , 'Country']
countries_2016 = data.loc[data['Year']==2016 , 'Country']
countries_2017 = data.loc[data['Year']==2017 , 'Country']
countries_2018 = data.loc[data['Year']==2018 , 'Country']
countries_2019 = data.loc[data['Year']==2019 , 'Country']

countries_2015 = set(countries_2015)
countries_2016 = set(countries_2016)
countries_2017 = set(countries_2017)
countries_2018 = set(countries_2018)
countries_2019 = set(countries_2019)

all_countries = countries_2015.union(countries_2016,countries_2017,countries_2018,countries_2019)
print('total countries ',len(all_countries))
all_common_countries = countries_2015.intersection(countries_2016,countries_2017,countries_2018,countries_2019)
print('common countries ',len(all_common_countries))
print('\n')
#common_countries
print('names of common countries','\n',all_common_countries)

#keepink only common countries
data = data[data['Country'].isin(all_common_countries)]
data = data.reset_index(drop=True)
#verifying 
#calculating missing in each year
countries_2015 = data.loc[data['Year']==2015 , 'Country']
countries_2016 = data.loc[data['Year']==2016 , 'Country']
countries_2017 = data.loc[data['Year']==2017 , 'Country']
countries_2018 = data.loc[data['Year']==2018 , 'Country']
countries_2019 = data.loc[data['Year']==2019 , 'Country']

countries_2015 = set(countries_2015)
countries_2016 = set(countries_2016)
countries_2017 = set(countries_2017)
countries_2018 = set(countries_2018)
countries_2019 = set(countries_2019)
print('\n')
all_countries = countries_2015.union(countries_2016,countries_2017,countries_2018,countries_2019)
print('total countries ',len(all_countries))
all_common_countries = countries_2015.intersection(countries_2016,countries_2017,countries_2018,countries_2019)
print('common countries ',len(all_common_countries))

#outliers do not need to be removed in this case sinc its real data some countries might have very good happiness score
```
